## MODEL SELECTION

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, log_loss
```

```python
# Split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
import skmultilearn
```

```python
from skmultilearn.problem_transform import BinaryRelevance
from skmultilearn.problem_transform import ClassifierChain
```

Coverting multi label into multi class classification

```python
def build_model(model,estimator,x_train,y_train,x_test,y_test):
    clf=estimator(model)
    clf.fit(x_train,y_train)
    clf_pred=clf.predict(x_test)
    acc=accuracy_score(y_test,clf_pred)
    ll=log_loss(y_test,clf_pred.toarray())
    result= {'accuracy':acc,'log_loss':ll}
    print(classification_report(y_test,clf_pred))
    return result
```

```python
B_r_mnb=build_model(MultinomialNB(),BinaryRelevance,x_train,y_train,x_test,y_test)
```

```python
B_r_mnb
```

```python
B_r_gnb=build_model(GaussianNB(),BinaryRelevance,x_train,y_train,x_test,y_test)
```

```python
B_r_gnb
```

```python
chain_model_Multi= build_model(MultinomialNB(),ClassifierChain,x_train,y_train,x_test,y_test)
```

```python
chain_model_Multi
```

```python
chain_model_Gau= build_model(GaussianNB(),ClassifierChain,x_train,y_train,x_test,y_test)
```

```python
chain_model_Gau
```

```python
chain_model_DTC= build_model(DecisionTreeClassifier(),ClassifierChain,x_train,y_train,x_test,y_test)
```

```python
chain_model_DTC
```

```python
chain_model_LR= build_model(LogisticRegression(),ClassifierChain,x_train,y_train,x_test,y_test)
```

```python
chain_model_LR
```

## HYPER PARAMETER

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'classifier':[MultinomialNB()],'classifier__alpha': [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]}]
Grid=GridSearchCV(ClassifierChain(),parameters,scoring='accuracy')
Grid.fit(x_train,y_train)
```

```python
print (Grid.best_params_, Grid.best_score_)
```

### Saving the best model

```python
clf=ClassifierChain(MultinomialNB(alpha=0.00001))
clf.fit(x_train,y_train)
clf_pred=clf.predict(x_test)
acc=accuracy_score(y_test,clf_pred)
ll=log_loss(y_test,clf_pred.toarray())
print({'accuracy':acc,'log_loss':ll})
```

```python
print(classification_report(y_test,clf_pred))
```
